## Gradient Descent for Linear Regression

```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,3,4,5])  # input 
y_train = np.array([1000,2000,3000,4000,5000])
```

```python
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('price in dollars')
plt.xlabel('size in square feet')
plt.legend()
plt.show()
```

```python
def cost(x, y, w, b):
    m = len(x)
    cost = 0
    
    for i in range(m):
        f = w * x[i] + b
        cost = cost + (f - y[i])**2
        
    total_cost = 1 / (2*m) * cost
    
    return total_cost
```

```python
def gradient(x, y, w, b):
    m = len(x)
    total_wg = 0
    total_bg = 0
    
    for i in range(m):
        f = w * x[i] + b
        w_g = 1 / m * (f - y[i]) * x[i]
        b_g = 1 / m * (f - y[i])
        total_wg = total_wg + w_g
        total_bg = total_bg + b_g
        
    return total_wg , total_bg
        
```

```python
def gradient_descent(x, y, w, b, iterations, alpha):
    
    for i in range(iterations):
        wg , bg = gradient(x, y, w, b)
        
        w = w - alpha * wg
        b = b - alpha * bg
        
        
        c = cost(x, y, w, b)
        print("fun_w {}, fun_b {}, cost {}, iterations {}".format(w, b, c, i))
     
    return w, b, c
    
    
    
```

```python
init_w = 0  # initial value of w
init_b = 0  # initial value of b
it = 2955   # no of iterations 
a = 0.1     # learning rate (alpha)
 
fun1, fun2, co =  gradient_descent(x_train, y_train, init_w, init_b, it, a)
```

```python
print("fun_w {}, fun_b {}, cost {} ".format(fun1, fun2, co ))
```

```python
print(f"(w,b) found by gradient descent: ({fun1:8.4f},{fun2:8.4f})") 
```

# Prediction

```python
print(f"1 sqft house prediction {fun1*1 + fun2:0.1f} Thousand dollars")
print(f"2 sqft house prediction {fun1*2 + fun2:0.1f} Thousand dollars")
print(f"3 sqft house prediction {fun1*3 + fun2:0.1f} Thousand dollars")
```
