## Implement the model for linear regression with one variable

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2,3,4,5])  # input 
y_train = np.array([1000,2000,3000,4000,5000]) # target

print(f"x_t = {x_train}")
print(f"y_t = {y_train}")
```

```python
m = len(x_train)
print(f"total training examples are : {m}")
```

```python
i = 0
x = x_train[i]
y = y_train[i]
print(f"(x^i and y^i) = ({x} , {y})")
```

### Plotting the data

```python
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.title("Housing Prices")
plt.ylabel('price in dollars')
plt.xlabel('size in square feet')
plt.legend()
plt.show()
```

### Model function

```python
# parameter
w = 1000  # weight
b = 0  # bias
```

```python
def model(x , w ,b):
    m = x.shape[0]     # m = len(x)
    arr = np.zeros(m)
    for i in range(m):
        arr[i] = w * x[i] + b  
        
    return arr
```

```python
fun = model(x_train,w,b)
```

```python
# Plot model prediction
plt.plot(x_train, fun, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price in dollars')
# Set the x-axis label
plt.xlabel('size in square feet')
plt.legend()
plt.show()
```
