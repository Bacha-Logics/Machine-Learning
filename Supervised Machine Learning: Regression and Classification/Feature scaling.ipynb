# Feature scaling

                      *** Before Normalization ***

```python
import copy, math
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[1, 5, 1, 45], [2, 3, 2, 40], [3, 2, 1, 35]]) 
y_train = np.array([1000, 2000, 3000])
```

```python
print(f"X Shape: {X_train.shape}")
print(X_train)
print(f"y Shape: {y_train.shape}")
print(y_train)
```

```python
def cost(x, y, w, b):
    m = x.shape[0]
    co = 0
    for i in range(m):
        fwb = np.dot(x[i],w) + b
        co = co + (fwb - y[i])**2
    co = co / (2 * m)
    return co
```

```python
def gradient(x, y, w, b):
    m,n = x.shape
    
    dw = np.zeros(n)
    db = 0
    
    for i in range(m):
        
        f =  (np.dot(x[i],w) + b) - y[i]
        
        for j in range(n):
            
            dw[j] = dw[j] + f * x[i,j]
        db = db + f 
        
    dw = dw / m
    db = db / m
    
    return dw,db
    
    
```

```python
def gradient_descent(x, y, w, b, iterations, alpha):
    
    
    for i in range(iterations):
        
        dw, db = gradient(x, y, w, b)
        
        w = w - alpha * dw
        b = b - alpha * db
    
        c = cost(x, y, w, b)
        print(f"Iteration {i:4d}: Cost {c:8.2f}   ")
    
    return w, b, c
```

```python
# initialize parameters
init_w = np.zeros_like(X_train[0])
init_b = 0
# some gradient descent settings
iterations = 25000
alpha = 0.001
# run gradient descent 
w_final, b_final,c = gradient_descent(X_train, y_train, init_w, init_b, iterations,  alpha)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
```

```python
print(f"prediction: {np.dot(X_train[0], w_final) + b_final:0.2f}, target value: {y_train[0]}")
```

```python
def predict(x, w, b): 
    p = np.dot(x, w) + b     
    return p    
```

```python
x_house = np.array([1, 5, 1, 45])
```

```python
res = predict(w_final,x_house,b_final)
```

```python
print(res)
```

```python
m = X_train.shape[0]
yp = np.zeros(m)
for i in range(m):
    yp[i] = np.dot(X_train[i], w_final) + b_final
    print(yp[i])
    # plot predictions and targets versus original features    
```

# z-score normalization

                           *** After Normalization ***

```python
def zscore_normalize_features(X):
    # find the mean of each column/feature
    mu     = np.mean(X, axis=0)                
    # find the standard deviation of each column/feature
    sigma  = np.std(X, axis=0)                  
    # element-wise, subtract mu for that column from each example, divide by std for that column
    X_norm = (X - mu) / sigma      

    return (X_norm, mu, sigma)
 
```

```python
X_norm, X_mu, X_sigma = zscore_normalize_features(X_train)
print(f"X_mu = {X_mu}, \nX_sigma = {X_sigma}")
```

```python
print(X_norm)
```

```python
def gradient_descent(x, y, w, b, iterations, alpha):
    
    
    for i in range(iterations):
        
        dw, db = gradient(x, y, w, b)
        
        w = w - alpha * dw
        b = b - alpha * db
    
        c = cost(x, y, w, b)
        print(f"Iteration {i:4d}: Cost {c:8.2f}   ")
    
    return w, b, c
```

```python
init_w = np.zeros_like(X_train[0])
init_b = 0
iterations = 3300
alpha = 0.003
```

```python
w_final, b_final,c = gradient_descent(X_norm, y_train, init_w, init_b, iterations,  alpha)
```

```python
x_house = np.array([1, 5, 1, 45])
x_house_norm = (x_house - X_mu) / X_sigma
x_house_predict = np.dot(x_house_norm, w_final) + b_final
print(f" predicted After Normalization House Price = ${x_house_predict:0.0f}")
```
