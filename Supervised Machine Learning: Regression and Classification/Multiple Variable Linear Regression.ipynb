# Multiple Variable Linear Regression

```python
import copy, math
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[10, 4, 4, 10], [20, 5, 5, 10], [30, 6, 6, 10]])    # [[1, 5, 1, 45], [2, 3, 2, 40], [3, 2, 1, 35]
y_train = np.array([1000, 2000, 3000])
```

```python
print(f"X Shape: {X_train.shape}")
print(X_train)
print(f"y Shape: {y_train.shape}")
print(y_train)
```

```python
def cost(x, y, w, b):
    m = x.shape[0]
    co = 0
    for i in range(m):
        fwb = np.dot(x[i],w) + b
        co = co + (fwb - y[i])**2
    co = co / (2 * m)
    return co
```

```python
def gradient(x, y, w, b):
    m,n = x.shape
    
    dw = np.zeros(n)
    db = 0
    
    for i in range(m):
        
        f =  (np.dot(x[i],w) + b) - y[i]
        
        for j in range(n):
            
            dw[j] = dw[j] + f * x[i,j]
        db = db + f 
        
    dw = dw / m
    db = db / m
    
    return dw,db
    
    
```

```python
def gradient_descent(x, y, w, b, iterations, alpha):
    
    
    for i in range(iterations):
        
        dw, db = gradient(x, y, w, b)
        
        w = w - alpha * dw
        b = b - alpha * db
    
        c = cost(x, y, w, b)
        print(f"Iteration {i:4d}: Cost {c:8.2f}   ")
    
    return w, b, c
```

```python
# initialize parameters
init_w = np.zeros_like(X_train[0])
init_b = 0
# some gradient descent settings
iterations = 850
alpha = 0.001
# run gradient descent 
w_final, b_final,c = gradient_descent(X_train, y_train, init_w, init_b, iterations, alpha)
print(f"b,w found by gradient descent: {b_final:0.2f},{w_final} ")
```

# Prediction

```python
print(f"prediction: {np.dot(X_train[0], w_final) + b_final:0.2f}, target value: {y_train[0]}")
```
